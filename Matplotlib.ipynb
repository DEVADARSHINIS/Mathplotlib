# Functional Methods in Matplotlib

**Matplotlib Library :** This is a visualization Library used to visualize data using plots,histogram etc.,

```python
# Load the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a list for X and Y
x = np.arange(0,10)
y = 2*x
```

```python
# print x
x
```

```python
# print y
y
```

**plot() :**This is used for **visual representation of Data.**

Here, we see that there is a line that displays what is the **type of plot()** and the **location where the image is saved.** kinda like a warning.

```python
# Create a plot for given x and y
plt.plot(x,y)
```

The type line with the location can be removed using **; or show().**

```python
# create a plot without the type displayed
plt.plot(x,y)
plt.show()
```

Here we have **title,axis labels,limit and to save the plot in an image format** on where the line should start.

```python
# create a plot without the type displayed
plt.plot(x,y);
# Create a title
plt.title('Data')
# Create labels for axis
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# Create limits
plt.xlim(0,8)
plt.ylim(0,18)
# save the image in png format
plt.savefig('Data.png')
```

### OOP Methods

```python
# Create a list
a = np.linspace(0,10,11)
b = a**4
```

```python
# Create a list
x = np.arange(0,10)
y = x*2
```

```python
# print the lists
x
```

```python
# print the lists
y
```

```python
# Create a blank plot
fig = plt.figure()
```

```python
# Create axes to that blank plot
fig = plt.figure()
# Large axes
axes_1 = fig.add_axes([0,0,1,1])
axes_1.plot(a,b);
# Functions
axes_1.set_title('Powers of 4 Graph')
axes_1.set_xlabel('X-Axis')
axes_1.set_ylabel('y-Axis')
# small axes
axes_2 = fig.add_axes([0.5,0.5,0.25,0.25])
axes_2.plot(x,y);
# Functions
axes_2.set_title('Linear Graph')
axes_2.set_xlabel('X-Axis')
axes_2.set_ylabel('y-Axis')
```

```python
# Create axes to that blank plot
fig = plt.figure()
# LARGE AXES
axes_1 = fig.add_axes([0,0,1,1])
axes_1.plot(a,b);
# Function to set title
axes_1.set_title('Powers of 4 Graph')
# Function to set label to x and y axis
axes_1.set_xlabel('X-Axis')
axes_1.set_ylabel('y-Axis')
# Functions to add limits in axes_1
axes_1.set_xlim(0,8)
axes_1.set_ylim(0,8000)


# SMALL AXES
axes_2 = fig.add_axes([0.5,0.5,0.25,0.25])
axes_2.plot(a,b);
# Functions to set title
axes_2.set_title('Power of 4 Zoomed In')
# Function to set label to x and y axis
axes_2.set_xlabel('X-Axis')
axes_2.set_ylabel('y-Axis')
# Functions to add limits in axes_1
axes_2.set_xlim(2,8)
axes_2.set_ylim(50,80)
# Save the plot
fig.savefig('new_file1.png',bbox_inches='tight')
```

**dpi :**This is improve the **resolution** of the plot.

```python
# Create
fig = plt.figure(figsize=(2,2),dpi=200)
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
fig.savefig('new_file.png',bbox_inches='tight')
```

# Subplot

```python
# Create subplots
fig,axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(a,b)
```

```python
# To find the type
type(axes)
```

```python
# Print the number of subplots
axes
```

```python
# Create subplots
fig,axes = plt.subplots(nrows = 2,ncols = 2)

axes[0][0].plot(x,y);
axes[1][1].plot(a,b);
```

```python
# Print the number of subplots
axes
```

```python
# Create subplots
fig,axes = plt.subplots(nrows = 3,ncols = 1)
# Using loop statements
for ax in axes:
  ax.plot(a,b)
# To make space between each graph
plt.tight_layout()
```

```python
# Create subplots
fig,axes = plt.subplots(nrows = 2,ncols = 2)

axes[0][0].plot(x,y);
axes[0][1].plot(a,b);
axes[1][0].plot(a,b);
axes[1][1].plot(x,y);

fig.subplots_adjust(wspace = 0.9,hspace = 0.5)
```

```python
# Create subplots
fig,axes = plt.subplots(nrows = 2,ncols = 2,dpi=200)
# Plot in the subplots
axes[0][0].plot(x,y);
# Functions
axes[0][0].set_title('Linear Graph');
axes[0][0].set_xlabel('X-Axis');
axes[0][0].set_ylabel('Y-Axis');


# Plot in the subplots
axes[0][1].plot(a,b);
axes[1][0].plot(a,b);
axes[1][1].plot(x,y);
# To get the enough space
fig.subplots_adjust(wspace = 0.5,hspace = 0.5)
# Create a super title for all the subplots
fig.suptitle('Subplots',fontsize=16);
# save the subplots
fig.savefig('subplots.png',bbox_inches='tight')
```

# Plot Styling

## Legend

```python
# Create a list
x = np.arange(0,11)
x
```

```python
# Create a plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,label='X vs X');
axes.plot(x,x**2,label='X vs X^2');
axes.legend()
```

```python
# Create a plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
# plot 2 graph
axes.plot(x,x,label='X vs X');
axes.plot(x,x**2,label='X vs X^2');
# this for the key
axes.legend(loc=(-0.2,1));
```

```python
# Create a plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
# plot 2 graph
axes.plot(x,x,label='X vs X');
axes.plot(x,x**2,label='X vs X^2');
# this for the key
axes.legend(loc=(1.1,1));
```

```python
# Create a list
x = np.linspace(0,11,10)
x
```

```python
# Create a plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,label='X vs X',color='purple',ls='--',lw=3);
axes.plot(x,x*2,label='X vs X*2',color='gold',linestyle='-',linewidth=3);
axes.legend();
```

```python
# Create a plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,label='X vs X',color='purple',ls='-.',lw=3,marker='o',ms=8);
axes.plot(x,x*2,label='X vs X*2',color='gold',linestyle='-',linewidth=3,marker='s',markersize=10);
axes.legend();
```

```python
# Create a plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,label='X vs X',color='purple',ls='-.',lw=3,marker='o',ms=8);
axes.plot(x,x*2,label='X vs X*2',color='gold',linestyle='-',linewidth=3,marker='s',markersize=10,markerfacecolor='red',markeredgecolor='purple',markeredgewidth=7);
axes.legend();
```

```python
# Create a plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x,label='X vs X',color='purple',ls='-.',lw=3,marker='o',ms=8);
axes.plot(x,x*2,label='X vs X*2',color='gold',linestyle='-',linewidth=3,marker='s',markersize=10,markerfacecolor='red',markeredgecolor='purple',markeredgewidth=3);
axes.legend();
```
